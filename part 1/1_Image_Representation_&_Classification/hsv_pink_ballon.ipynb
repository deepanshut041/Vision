```python
import numpy as np
import cv2
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

PATH = "images/water_balloons.jpg"
image = cv2.imread(PATH)
print('This image is of type: ', type(image),
     ' With dimensions :', image.shape)
image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

### Plot Color channels

```python
# RGB Channels
r = image_copy[:, :, 0]
g = image_copy[:, :, 1]
b = image_copy[:, :, 2]
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(30, 10))

ax1.set_title('Red')
ax1.imshow(r, cmap='gray')

ax2.set_title('Green')
ax2.imshow(g, cmap='gray')

ax3.set_title('Blue')
ax3.imshow(b, cmap='gray')
```

```python
image_hsv = np.copy(image_copy)
image_hsv = cv2.cvtColor(image_hsv, cv2.COLOR_RGB2HSV)
plt.imshow(image_hsv, cmap='hsv')
```

```python
h = image_hsv[:, :, 0]
s = image_hsv[:, :, 1]
v = image_hsv[:, :, 2]
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(30, 10))

ax1.set_title('Hue')
ax1.imshow(h, cmap='gray')

ax2.set_title('Saturation')
ax2.imshow(s, cmap='gray')

ax3.set_title('Value')
ax3.imshow(v, cmap='gray')
```

```python
# Defining RGB selection criteria
lower_pink = np.array([180, 0, 100])
upper_pink = np.array([255, 255, 230])

# Define our hsv selection criteria
lower_hue = np.array([160, 0, 0])
upper_hue = np.array([180, 255, 255])
```

```python
# Masking image
mask_rbg = cv2.inRange(image_copy, lower_pink, upper_pink)
mask_hsv = cv2.inRange(image_hsv, lower_hue, upper_hue)

masked_rbg = np.copy(image_copy)
masked_hsv = np.copy(image_hsv)

masked_rbg[mask_rbg == 0] = [0, 0, 0]
masked_hsv[mask_hsv == 0] = [0, 0, 0]

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(30, 10))

ax1.set_title('Rgb Method')
ax1.imshow(masked_rbg)

ax2.set_title('Hsv Method')
ax2.imshow(masked_hsv)
```
