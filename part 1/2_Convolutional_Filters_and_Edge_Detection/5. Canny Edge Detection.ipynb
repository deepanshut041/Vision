```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

# Read in the image
image = cv2.imread('images/brain_MR.jpg')

# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
# Convert the image to grayscale for processing
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
blur = cv2.GaussianBlur(gray, (5, 5), 0)

# Display the images
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('Gray')
ax1.imshow(gray, cmap='gray')

ax2.set_title('Blur Gray')
ax2.imshow(blur, cmap='gray')
```

### Implement Canny edge detection

```python
# Try Canny using "wide" and "tight" thresholds

wide = cv2.Canny(gray, 30, 100)
tight = cv2.Canny(gray, 200, 240)

wide_blur = cv2.Canny(blur, 30, 100)
tight_blur = cv2.Canny(blur, 200, 240)
    
# Display the images
f, plots = plt.subplots(2, 2, figsize=(20,20))

plots[0][0].set_title('wide')
plots[0][0].imshow(wide, cmap='gray')

plots[0][1].set_title('tight')
plots[0][1].imshow(tight, cmap='gray')

plots[1][0].set_title('wide blur')
plots[1][0].imshow(wide_blur, cmap='gray')

plots[1][1].set_title('tight blur')
plots[1][1].imshow(tight_blur, cmap='gray')
```
